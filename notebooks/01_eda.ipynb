## 1. 데이터 로드 및 기본 검증

```python
# data/raw/ 폴더에 CSV 파일이 있어야 합니다
# kaggle datasets download -d algozee/teenager-menthal-healy
# unzip *.zip -d ../data/raw/

import os
raw_files = [f for f in os.listdir('../data/raw') if f.endswith('.csv')]
print('발견된 파일:', raw_files)

df = load_raw_data(raw_files[0])
print(f'\n데이터 크기: {df.shape}')
df.head()
```
